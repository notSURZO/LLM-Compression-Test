```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.retrievers.document_compressors import LLMChainExtractor
from langchain.retrievers import ContextualCompressionRetriever
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.docstore.document import Document
from langchain.schema import Document as LangChainDocument
import os

# --- Set up Gemini 1.5 Flash ---
GOOGLE_API_KEY = 'AIzaSyAGLNHPCQ2VL2kULFP59e1O0YMAOkhYy5Y'  # Replace this
llm = ChatGoogleGenerativeAI(
    model="models/gemini-1.5-flash",
    google_api_key=GOOGLE_API_KEY,
    temperature=0.0,
    max_output_tokens=512
)

# --- Step 1: Sample Documents ---
docs = [
    Document(page_content="Mitochondria are the powerhouse of the cell. They convert glucose into ATP. Also, the cell has ribosomes that synthesize proteins. The nucleus controls gene expression"),
    Document(page_content="Photosynthesis occurs in chloroplasts and converts light energy into chemical energy."),
    Document(page_content="The nucleus contains the cell‚Äôs genetic material and coordinates cell activities."),
]

# --- Step 2: Setup Compression ---
compressor = LLMChainExtractor.from_llm(llm)

# --- Step 3: Compress the Documents using a Query ---
query = "What is the role of mitochondria?"

# This simulates what LangChain does internally
compressed_docs = compressor.compress_documents(docs, query=query)

# --- Step 4: Print the Output ---
print("üîç Original Documents:\n")
for i, doc in enumerate(docs, 1):
    print(f"[{i}] {doc.page_content}\n")

print("\nüìé Compressed Documents:\n")
for i, doc in enumerate(compressed_docs, 1):
    print(f"[{i}] {doc.page_content}\n")
```
